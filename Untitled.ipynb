```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hardcoded example data
X = np.array([[1, 2, 1], [1, 3, 2], [1, 4, 3]])  # Example predictor matrix (3x3)
y = np.array([[3, 4, 5], [4, 5, 6], [5, 6, 7]])  # Example response values (3x3)

# Compute the projection of y onto the subspace V spanned by the columns of X
Xt_X_inv = np.linalg.inv(X.T @ X)
beta_hat = Xt_X_inv @ X.T @ y
y_hat = X @ beta_hat

# Plot the data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(y[:, 0], y[:, 1], y[:, 2], label='y', color='blue')
ax.scatter(y_hat[:, 0], y_hat[:, 1], y_hat[:, 2], label='y_hat', marker='x', color='red')
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')
ax.set_title('Projection of y onto subspace V')
ax.legend()
plt.show()
```
